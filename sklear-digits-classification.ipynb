```python
from sklearn import datasets
from sklearn.cross_validation import cross_val_score
from sklearn import svm, metrics

from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import RandomizedPCA
from sklearn.ensemble import RandomForestClassifier

import pylab as pl
import numpy as np
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
print digits.DESCR
```

```python
X = digits.data
y = digits.target
```

```python
n_samples, n_features = X.shape
```

```python
pl.gray()
pl.rcParams['figure.figsize'] = 12, 8.5
pl.rcParams['axes.grid'] = True

for i, j in enumerate(np.random.permutation(X.shape[0])[:7]):
    pl.subplot(2, 7, (i + 1))
    pl.imshow(X[j].reshape((8, 8)), interpolation='nearest')
    pl.title("true class: %d" % y[j])
    pl.xticks(()), pl.yticks(())
    
for i, j in enumerate(np.random.permutation(X.shape[0])[:7]):
    pl.subplot(2, 7, (7 + i))
    pl.imshow(X[j].reshape((8, 8)), interpolation='nearest')
    pl.title("true class: %d" % y[j])
    pl.xticks(()), pl.yticks(())    
```

```python
X_pca = RandomizedPCA(n_components=2).fit_transform(X)
```

```python
X_pca.shape
```

```python
# Some nice default configuration for plots
pl.rcParams['figure.figsize'] = 10, 7.5
pl.rcParams['axes.grid'] = True
pl.gray()

from itertools import cycle
color = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i, c in zip(np.unique(y), cycle(color)):
    pl.scatter(X_pca[y==i, 0], X_pca[y==i, 1], c=c, label=str(i), alpha=.8)
    pl.legend(loc='upper right')
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33, random_state=0)
svc = svm.SVC(kernel='rbf').fit(X_train, y_train)
svc_score = svc.score(X_test, y_test)
print svc_score
```

```python
svc
```

```python
svc2 = svm.SVC(kernel='rbf', C=100, gamma=.001).fit(X_train, y_train)
svc2_score = svc2.score(X_test, y_test)
print svc2_score
```
